```python
from tokenizer import BPETokenizer 
```

```python
import os
from pathlib import Path

dataset_dir = Path("dataset/harry-potter-books") 

os.listdir(dataset_dir)
```

```python
train_corpus = ""

for book in os.listdir(dataset_dir):
    book_path = dataset_dir / book
    with open(book_path, 'r') as f:
        train_corpus += f.read()

len(train_corpus)
```

```python
enc = BPETokenizer()
enc.train(train_corpus, 1024, verbose=True)
```

```python
enc.stoi
```

```python
with open("hp_tokenizer_dict.json", "w") as f:
    import json
    json.dump(enc.stoi, f, indent=4)
```
